```python
best_department = data.pivot_table(values = "MonthlyIncome" , index = "Department" , aggfunc = "sum")
best_department = best_department.sort_values(by=['MonthlyIncome'], ascending=False).head(15)
best_department.reset_index(inplace=True) # we will have to reset the index to add the customer name into dataframe 
best_department['MonthlyIncome']=best_department['MonthlyIncome'].round(2)
best_department
```

```python
fig = plt.figure(figsize = (10, 10))
 
# creating the bar plot
plt.bar(best_department['Department'],best_department['MonthlyIncome'] ,width = 0.4)
 
plt.xlabel("Best Department",fontsize=15)
plt.ylabel("Total Monthly Income",fontsize=15)
plt.xticks(fontsize=12, rotation=90)
plt.yticks(fontsize=12)

for i , v in enumerate(best_department['MonthlyIncome']):
    plt.text(i, v/2 , str(v) , ha='center' , va='center' , rotation=90 , color='white')
plt.show()
```
